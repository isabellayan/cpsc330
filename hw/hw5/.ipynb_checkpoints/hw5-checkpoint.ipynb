Showing the features again above, as there're no nulls, no imputation is required. 

Using the table from hw3: 
| Feature | Transformation | Explanation
| --- | ----------- | ----- |
| ID | drop | This is used to identify each row of data. Since this will probably not be contributing to the prediction, I'll drop this feature. |
| LIMIT_BAL | scaling | This shows the amount of given credit, a numeric feature ranging from 10000 to 1000000. Since the mean and standard deviation is quite different from other numeric features, I would apply scaling. |
| SEX | one-hot encoding with "binary=True" | This is the gender of the people in the dataset with only two categories. Since there's no ordinality here, we should perform one-hot encoding with "binary=True". Although the column is already encoded with 1 and 2, I'll encode it again so it becomes 0 and 1. |
| EDUCATION | ordinal encoding | This is showing the different education levels of the people in the dataset. Since categories except others have ordinality, I would use ordinal encoding here and treat others as the lowest level since there's no information on their education level. Although the numbers are already encoded in the correct order, I'll encode them again to make them start from 0. |
| MARRIAGE | one-hot encoding | This is showing whether the clients are married, single, or others. Since there's no ordinality here, I'll use one-hot encoding to distinguish between the values. |
| AGE | scaling | This is the age in years of the people in the dataset, ranging from 21 to 79. The reason for scaling is the same with the feature LIMIT_BAL. |
| PAY | passthrough | This is the repayment status of each client for each month. Although there's ordinality here as pay duly is better than delay for one month and so on, the result is already encoded in the correct order so I'll make it passthorugh here. |
| BILL_AMT | scaling | This is the amount of bill statements in dollar, and the range varies greatly from negative to positive. The reason for scaling is the same with the feature LIMIT_BAL. |
| PAY_AMT | scaling | This is the amount of previous payments in dollar, and the range varies greatly from 0 to a large positive number. The reason for scaling is the same with the feature LIMIT_BAL. |

```python
# using code from hw3

results_dict = {}  # dictionary to store all the results
```

```python
def mean_std_cross_val_scores(model, X_train, y_train, **kwargs):
    """
    Returns mean and std of cross validation

    Parameters
    ----------
    model :
        scikit-learn model
    X_train : numpy array or pandas DataFrame
        X in the training data
    y_train :
        y in the training data

    Returns
    ----------
        pandas Series with mean scores from cross_validation
    """

    scores = cross_validate(model, X_train, y_train, **kwargs)

    mean_scores = pd.DataFrame(scores).mean()
    std_scores = pd.DataFrame(scores).std()
    out_col = []

    for i in range(len(mean_scores)):
        out_col.append((f"%0.3f (+/- %0.3f)" % (mean_scores[i], std_scores[i])))

    return pd.Series(data=out_col, index=mean_scores.index)
```

```python
import warnings
from sklearn.dummy import DummyClassifier
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.model_selection import cross_val_score, cross_validate

warnings.simplefilter('ignore')
scoring = [
    "accuracy",
    "f1",
    "recall",
    "precision",
]

dummy = DummyClassifier(random_state = 123)
pipe = make_pipeline(preprocessor, dummy)
results_dict["dummy"] = mean_std_cross_val_scores(
    pipe, X_train, y_train, cv=5, return_train_score=True, scoring=scoring
)
results_df = pd.DataFrame(results_dict).T
results_df
```

The accuracy of the dummy classifier is 0.777, meaning 77.7% of the training data belongs to the same class. This is same as the result from EDA.

```python
from sklearn.linear_model import LogisticRegression

lr_pipe = make_pipeline(preprocessor, LogisticRegression(max_iter=1000))
scores = cross_validate(lr_pipe, X_train, y_train, return_train_score=True)
pd.DataFrame(scores)
```

```python
# some code from lecture 7
cv_dict = {}

scores_dict = {
    "C": 10.0 ** np.arange(-4, 6, 1),
}
for C in scores_dict["C"]:
    lr_pipe_cv = make_pipeline(preprocessor, LogisticRegression(C=C, max_iter=1000))
    cv_dict[str(C)] = mean_std_cross_val_scores(
        lr_pipe_cv, X_train, y_train, cv=5, return_train_score=True
    )

results_df = pd.DataFrame(cv_dict).T
results_df
```

```python
scoring = [
    "accuracy",
    "f1",
    "recall",
    "precision",
]

lr_pipe_result = make_pipeline(preprocessor, LogisticRegression(C=1, max_iter=1000))
results_dict["Logistic Regression"] = mean_std_cross_val_scores(
    lr_pipe_result, X_train, y_train, cv=5, return_train_score=True, scoring=scoring
)
results_df = pd.DataFrame(results_dict).T
results_df
```

```python
lr_pipe_result.fit(X_train, y_train)
lr_pipe_result['logisticregression'].classes_
```

```python
# from lecture 9
thresholds = np.arange(0.0, 1.0, 0.1)
thresholds
```

```python
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score

pr_dict = {"threshold": [], "precision": [], "recall": [], "f1 score": []}
for threshold in thresholds:
    preds = lr_pipe_result.predict_proba(X_train)[:, 1] > threshold
    pr_dict["threshold"].append(threshold)
    pr_dict["precision"].append(precision_score(y_train, preds))
    pr_dict["recall"].append(recall_score(y_train, preds))
    pr_dict["f1 score"].append(f1_score(y_train, preds))
pd.DataFrame(pr_dict)   
```

Logistic Regression is a linear model for classification, so I used the model here. Both the cross validation score and the train score increases as C increases. Although a bigger C might lead to overfitting but it isn't seen in this case. The two scores are the same and the highest for all C values >= 1, so I'll choose the lowest C where C=1 in this case. Both the training score and the cross validation score is higher than dummy classifier. However, the test recall score is really low here with only 0.327, meaning we miss most of the default payment clients. Looking at the precision/recall tradeoff, the model seems to be performing the best when threshold = 0.2, where the f1 score is the highest. Using this score, the model's recall score is able to increase to around 0.6. See what non-linear models can do.
